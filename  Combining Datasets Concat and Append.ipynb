# Combining Datasets: Concat and Append


```python
#importing pandas library and creating a DataFrame 
import pandas as pd
df1=pd.DataFrame({'A':[1,2,3,4], 'B':[5,6,7,8]})
df2=pd.DataFrame({'A':[9,10,11,12],'B':[14,15,16,17]})

print(df1);print(df2)
```

```python
# concatenation of df1 and df2 
pd.concat([df1,df2])
```

```python
# convering the df2 indices into sequence series 
pd.concat([df1,df2],ignore_index=True)
```

```python
# concatenation of df1 and df2 by columns wise  using axis =1 or axis = columns 

pd.concat([df1,df2],axis=1)
```

```python
# creating one more rows 
df3=pd.DataFrame({'C':[18,19,20,21]})
df3
```

```python
# concatenation of df3 into df1 df2
pd.concat([df1,df2,df3],ignore_index=True)
```

```python
# concatenation of df3 into df1 df2 by columns wise 
pd.concat([df1,df2,df3],axis=1)
```

```python
# acessing the column A
acess=pd.concat([df1,df2,df3],axis=1)
acess['A']
```

```python
# by doing fancing indexing we can acessing more column index
acess[['A','B']]
```

```python
# adding the new  column 
acess['D']=[22,23,24,25]
acess
```

```python
acess=pd.concat([df1,df2,df3],axis=0)
acess
```

```python
# adding  multiindex key  to specify a label for the data sources
pd.concat([df1,df2,df3],keys=[' sec A',' sec B','sec C'])
```

# Catching the repeats as an error


```python
try:
    pd.concat([df1,df2,df3], verify_integrity=True)
except ValueError as e:
    print("ValueError:", e)
```

# Concatenate rows with different columns


```python
# now we are going to perform  join parameter 
pd.concat([df1,df2])
```

```python
# perfoming the join parameter 
pd.concat([df1,df2],join='outer')
```

```python
# now perfoming join='inner ' parameter  which gives all the commans rows and columns 
pd.concat([df1,df2],join='inner',ignore_index=True)
```

```python
# performimg the join prameter by coulmns wise
pd.concat([df1,df2],join='inner',axis=1)
```
