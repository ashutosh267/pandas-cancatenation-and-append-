# Combining Datasets: Merge and Join


```python
Categories of Joins
1.One to one :-When a row in a table is related to only one row in another table and vice versa,we say that is a one to one relationship 
2.Many to one :-This is where a row from many table can have one matching rows in another table this relationship is defined as a one to many relationship.
3.Many to many:- A row from one table can have multiple matching rows in another table, and a row in the other table can also have multiple matching rows in the first table this relationship is defined as a many to many relationship
    
```

# One to one 
When a row in a table is related to only one row in another table and vice versa,we say that is a one to one relationship 


```python
import pandas as pd 
```

```python
# creating a dataframe to perform one to one 

df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group':['Account','Engineering','Engineering','Hr']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
```

```python
df1
```

```python
df2
```

```python
# merging the datafarme i.e. df1 to df2
df3=pd.merge(df1,df2)
df3
```

# Outer join | Full outer join 

In Outer join | Full outer join  basically it will give you all the element in the datframe after mergeing dataframe like in the set we known as union 


```python
# perfoming a outer join by passing through how parameter
pd.merge(df1,df2,how='outer')
```

## Cross Join

 In cross join it will multiply one element from your dataframe with the all  element in your other dataframe like cartesian product or cross multiplication 

```python
# perfoming a Right join | Right outer join  by passing through how parameter
pd.merge(df1,df2,how='cross')
```


## Overlapping Column Names: The suffixes Keyword

In a case where your two input DataFrames have conflicting column names in thos scenarios we can make use of suffixes bydefault 
the merge function automatically appends a suffix _x or _y to make the output columns unique.

basically by join left and right dataframe to identyfing the dataframe from which one belong to which one dataframe like left and right dtaframe  so we can make use of suffix to give our own  suffix name we can see in the below task 

```python
# performing the suffixes keyword to  df1 and df2  
pd.merge(df1,df2 , on ='FellowshipID',how ='outer', suffixes=('_df1 left','_df2 right'))
```

## Indicator keyword

Indicator keyword basically shows what type of merging is done it create one column in df and show merge type 

```python
#performing the indicator keywords  using how parameter 
pd.merge(df1,df2,how='outer', indicator=True)
```

```python
#performing the indicator keywords  using how parameter 
pd.merge(df1,df2,how='inner', indicator=True)
```

```python
#performing the indicator keywords  using how parameter 
pd.merge(df1,df2,how='right', indicator=True)
```

```python
#performing the indicator keywords  using how parameter 
pd.merge(df1,df2,how='left', indicator=True)
```

```python
#performing the indicator keywords  using how parameter 
pd.merge(df1,df2,how='cross', indicator=True)
```
