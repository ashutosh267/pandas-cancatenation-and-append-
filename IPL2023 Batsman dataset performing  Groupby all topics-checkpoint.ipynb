# GroupBy: Split, Apply, Combine


Groupby() is used to make groups of any column in our dataset by passing the column name in groupby ().

The groupby function basically work on  these three  Split, Apply, Combine .

Split:-The split step involves breaking up and grouping a DataFrame depending on the
value of the specified key.

Apply:-The apply step involves computing some function, usually an aggregate, transfor‚Äê
mation, or filtering, within the individual groups.

Combine:-The combine step merges the results of these operations into an output array.

# Aggregate() 

As we already know the  gorupby aggregate () like  sum(),mean() and  etc.

In groupby aggregate () method allows for even more flexibility. It can take
a string, a function, or a list thereof, and compute all the aggregates at once place.

Groupby aggregate () methed also allows to   pass a dictionary  to perform aggregate ().


```python
# performing the  min,max,mean by passing  through string etc  using aggregate () 

ipl1.aggregate(['min','max','mean']).head()
```

```python
# performing the  min,max,mean by passing  through list etc  using aggregate () 
ipl1.aggregate([min,max,sum]).head()
```

```python
# performing the  min,max,mean by passing  through string,list etc  using aggregate () 
ipl1.aggregate(['min',max,sum]).head()
```

```python
ipl1.aggregate(['min',max,mean]).head()
```

```python
# performing the  min,max,mean by passing  through string,list etc  using aggregate ()
import numpy as np
ipl1.aggregate(['min',max,np.mean,np.median]).head()
```

# Groupby aggregate () methed also allows to   pass a dictionary  to perform aggregate ().


```python
# passing the dictionary in aggregate () to perform sum artmetic operations
ipl1.aggregate({'Run':'mean',
                'Ball':'sum',
               'Fours':'median',
                'Sixes':'max',}).head()
```

```python
# apply aggregate to the single batsman player through list

ipl1.get_group('Virat Kohli').aggregate(['min','max','mean']).head()
```

```python
# apply aggregate to the single batsman player through dictionary 

ipl1.get_group('Virat Kohli').aggregate({'Run':'mean',
                                         'Ball':'sum',
                                         'Fours':'median',
                                         'Sixes':'max'}).head()
```

```python
# by passing the dictionary 
ipl1.get_group('Virat Kohli').aggregate({'Run':['mean','max',sum,'median','mean']})
                                
```

# Filtering 

Filtering. A filtering operation allows you to drop data based on the group properties.

Basically it drop the False value and return the True boolean value only.
 
For example suppose  we are performing the addition task of two numbers and want to filter according to its value .
 a=4+5=9 and b=5+3=8 by adding these the value of is a=9 ,b=8
 
if the we filter according to which  is greater then  a or b by applying filter (), then it will only filter which one is greater number simply it will filter a value of a variable  that is 9 and drop the value of b that 8 which is not greater thrn a
thats how filter works


```python
# showing  the dataset
ipl1.head()
```

```python
# seeing which player has scored  run less then 1 run.
def oye(d):
    return d['Run'].sum()<1
```

```python
# passing the oye function into the filter ()

w=ipl1.filter(oye)
w
```

```python
# taking the  sum of the  ball by 
def hey(t):
     return t['Ball'].sum()>120
    
```

```python
# passing the hey function into the filter ()
d=ipl1.filter(hey)
d
```

```python
# seeing which batsman has delevered 69 fours in the 2023 ipl season and in which match_no. has score how much fours
def see(g):
    return g['Fours'].sum()==69
```

```python
dsa=ipl1.filter(see)
dsa
```

```python
# seeing the total no.  of fours
dsa['Fours'].sum()
```

```python
# seeing the maximum fours delevere by batsman from the above dsa variable
dsa['Fours'].max()
```

```python
# fetching all the sixes beaten by faf 
sa=ipl1.filter(lambda x: x['Sixes'].sum()>35)
sa

```

```python
ipl1['Sixes'].sum().sort_values(ascending=False)
```

```python


# seeing both data at once place only 
display(ipl1['Sixes'].sum().sort_values(ascending=False),sa)
```

# Transformation/ Transform()



In transformation  if you want to get a new value for each original row, use tranform()  function .

Basically it put your desired output into your  each original dataset  input  rows  thats how it work like.

As name suggesting it will tranformation can return some transformed version of the full data to recombine and  the output is the same shape as the input.

For example suppose you have small data set of two rows and two columns and have performed sum() aggregation get your desired output  and you want to fill these output  value in the same dataframe without reducing it shape that is rows and columns in the dataset  then you can easily fill by  using these transfrom() methods

```python
# calling the ipl1 dataset 
ipl1.head()
```

```python
# fetching all records of the one batsman  Devon Conway
ipl1.get_group('Devon Conway')
```

```python
# fetching all records of the one batsman  Devon Conway and taking its mean()
ipl1.get_group('Devon Conway').mean()
```

```python
# performing the mean of  using transpose () at ipl1
ipl1.transform(lambda x: x.mean())
```

```python
# performing the max() of  using transpose () at ipl1 
ipl1.transform(lambda x: x.max())
```

# The apply() method.

In these method basically we can apply our user define function or lamda and pass it to apply method().

The function should take a DataFrame, and return either a Pandas
object (e.g., DataFrame, Series) or a scalar; the combine operation will be tailored to
the type of output returned.

```python
# calling the ipl1 datasset 
ipl1.head()
```

```python
# applying() the function to get ball.sum
def he(x):
    x['Ball'].sum()
    return x
```

```python
ipl1.apply(he)
```
