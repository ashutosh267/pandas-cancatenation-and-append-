## Full join or outer join
Basicaly in fulljoin it takes all the element from both the dataset and common element of dataset also.
 
For example: suppose A and B are two data set and we perform fulljoin or outer join then it would look like (A union B) along with its common element that is (A interction B)

Now  we can do addition of these two union dataset and then minus the common element i.e. intersection 

Basicaly it look like:- (A + B) -(A intersection B)

```python
# performing outer join
outerjoin=pd.merge(order,returns,on='Order ID',how='outer',indicator=True)
outerjoin.head()
```

## Cross joint

In cross joint it multiplying all the element from dataframe one to other dataframe  when its merge.

In cross joint we can not use on keyword because it use cartesian product.

```python
# performing cross join
cross=pd.merge(order,returns, how='cross',indicator=True)
cross.head(20)
```

```python
# taking 5 rows and columns from both the dataset to easily perform cross join and  anyone can easily understand 
cross=pd.merge(order[:5],returns[:5],how='cross', indicator=True)
cross
```

```python
# we  use on keyword we can get  an error because in cross join we can not use on, right_on, left_on and right_index, left_index we can see in below
cross=pd.merge(order[:5],returns[:5],on='Order ID' ,how='cross', indicator=True)
cross
```

```python
cross._merge.value_counts()
```

## Left_index and right_index

As we can merge data according to the index also using this parameter.

```python
# reading the data from excel sheet   using clipboard
emp=pd.read_clipboard()
emp
```

```python
# reading second dataset that is salary using clipboard
sal=pd.read_clipboard()
sal
```

```python
# setting the index of emp dataset
sal.set_index('EmpID',inplace=True)
sal
```

```python
# performing the merge  using on keyword
index_wise=pd.merge(emp,sal,on='EmpID')
index_wise
```

```python
# performing the index wise merge using left_index and right_index bydefault join in inner 

index_wise=pd.merge(emp,sal,left_index=True,right_index=True)
index_wise
```
