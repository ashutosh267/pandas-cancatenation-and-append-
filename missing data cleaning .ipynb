# Handling missing value 

Missing values come in a dataset for various reason one might be while doing survey people did not fill the survey form properly etc .

We are here to deal with missing values 

```python
# importing the library 

import pandas as pd 
import numpy as np
```

```python
# creating a dataframe  with nan values 

data = pd.DataFrame({'Names':['Marcus','Rocky','Micko','Loffy',np.nan],
                    'Marks':[100,24,33,45,50],
                    'Ages':[10,13,None,np.nan,None]})
data
```

```python
type(data)
```

# isnull()

isnull() methods  Generate a Boolean mask indicating missing values

When there is nan value its retun True  and if there  is  no null values it return False boolean 

Basically it gives nan values as true in the df or series in boolean mask indication 

```python
# checking the null values in dataset 

check=data.isnull()
check
```

```python
# seeing the data at one place using display() method 

display(data , check)
```

# Sometimes we use compliments ~ to change from notnull() to isnull() to perfrom the task  and its denoted by ~
belows some examples


```python
com = data[~data.isnull()] # as we have applied compliment(~) it can change from isnull() to notnull() method
com
```

# dropna() Dropping null values

In Dropping null values it drops all the null values in the dataset which have null values

In droppna() method we can drop one rows or columns we can not drop one null values if we drop one row or columns then we might loss huge amount of data to deal with this if 95% of data se null values then we can delet and when data less than 5% then we can fill it by fiilna method.

In dropna() method there is  how ='any' paratmeter bydefault any parameter can drop all the column or row which has one null values  and how= 'all', will drop if in th ecolumns or rows has all null values.


In dropna()method there is threshold i.e thresh parameter in which we can specify the  minimum number
of non-null values for the row/column to be kept:
    

Dropping null values is completely based on the your how much  null values in your data if its is less than 5% then we can will it by numerical values or by mean 
ann if the data is in categorical i.e mixed of integer and strings then we can fill with most times occuring  number  when we counts.

```python
# performing the dropna() on the dataset 

remove = data.dropna() # it will drop all the null value  bydefault  rowise i.e. axis= 'row'
remove
```

```python
# performing the dropna () by columnwise 

col = data.dropna( axis='columns') # it will drop the null value  from column wise by passing through the arguments  in   dropna()
col
```

```python
# seeing the data df and col variable togather 

display(data,col)
```

```python
# making one more columns in data df  and putting the nan values
data['Rank'] =  np.nan
data
```

```python
# making the how = 'all' parameter to drop who has all null values, bydefault it is 'any'

all = data.dropna(how='all',axis=1) # how = all parameter will dropp who have all null values in th ecolumns
all
```

# Thresh

the thresh parameter lets you specify a minimum number of non-null values for the row/column to be kept:

```python
# perfoming the  thresh parameter 

thre = data.dropna(axis=0, thresh=2)
thre
```

```python
display(data,thre)
```

# Filling null values

Filling null values by using the fillna() methods sometimes its required to fill the values to calculate the numerical values 

Filling null values i.e . fillna() has  two method to fill the null values i.e to ffill it stand fro forward fill and bfill it stand for backward fill

```python
# performing the fillna() method to fill the data 

filling = data.fillna(0) # it fill nan value with zeros because we have passed zeros in the fillna() method 
filling
```

```python
# perfoming the fillna method to fill forward fill
okay = data.fillna(method = 'ffill') # it will fill values in forward manner
okay
```

```python
# changing the values of columns ranks at position 3 from columnwise and by rowise it is in the 2
data.iloc[2,3]=110
data
```

```python
# backward fill
hey1 = data.fillna(method = 'bfill', axis=0) # it will fill values in backward manner
hey1
```

```python
display(data,hey1)
```

```python
# backward filling by colunwise
good = data.fillna(method='bfill',axis=1)
good
```
