# Filetring the data based on condition using bitwise operators i.e and ,not, or xor etc

& (and) true & true = True  else all are false

| (or) Flase | False =False else all are true

~ (not) it's is compliment of of any operatin if you perform and operation using not operator it give output as or operaters


 Syntax:-  df[(df['column']== '  ') & (df['column']=='  ')]

```python
# now filter the data based on  state 

states = df[(df['State']== 'New York') & (df['City']=='New York')]
states.head()
```

```python
# counting the values after performing the filtering on states

states.State.count()
```

```python
# counting the values after performing the filtering on city 

states.City.count()
```

```python
states.State.value_counts()
```

```python
states.City.value_counts()
```

```python
# filtering the retailer based on sales method

filt = df[(df['Retailer']=='Foot Locker') & (df['Sales Method']=='Outlet')]
filt
```

```python
 # checing the mississippi is in state or not 
    
df[df['State'].isin(['Mississippi'])]
```

```python
# using or bitwise operatare

ret =  df[(df['Retailer']=='Walmart') | (df['Retailer']=='Amazon')]
ret
```

# series.isin() using this fuction 

```python
cities = ['Jackson','Houston','New York','Birmingham','Las Vegas' , 'Denver','Los Angeles','San Francisco']
citiyis = df[(df['City'].isin(cities)) & (df['Invoice Date']=='2020')]
citiyis
```

```python
plt.style.use('fivethirtyeight')
citiyis.plot(kind='line',figsize=(10,5))
plt.show()
```

# not ~

```python
cities = ['Jackson','Houston','New York','Birmingham','Las Vegas' , 'Denver','Los Angeles','San Francisco']
citiyis = df[~(df['City'].isin(cities)) & (df['Invoice Date']=='2020')]
citiyis
```

# Subquery()

subquery is used to avoid hard coding and which gives futur proof of the calculation

```python
# taking mean() 

sub = df['Price per Unit'].max()
sub
```

```python
# filtering the data using above timestamp

fil = df[(df['Price per Unit']==sub)]
fil
```

```python
fil.plot(kind='bar', title='Maximum Price Per Unit')
plt.show()
```

```python
fil.plot.pie( y ='Price per Unit', title='Maximum Price Per Unit')
plt.show()
```
