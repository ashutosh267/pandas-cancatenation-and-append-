# Resampling and Frequency Conversion

Resampling refers to the process of converting a time series from one frequency to another.

To use resampling we can apply aggregate method.



syntax:-

DataFrame.resample(rule, axis=_NoDefault.no_default, closed=None, label=None, convention='start', kind=None, on=None, level=None, origin='start_day', offset=None, group_keys=False)[source]




```python
# perfoming the resampling() to see minimum stock price on a year end

res = adaniports.resample(rule='A').min()
res
```

```python
# perfoming the resampling() to see maximum stock price on a year end 

res = adaniports.resample(rule='A').max()
res
```

```python
# perfoming the resampling() to see maximum stock price on a year end 

res1 = adaniports.resample(rule='A').max()
res1
```

```python
# perfoming the resampling() to see monthend average or mean price 

monthend = adaniports.resample(rule='M').mean()
monthend
```

```python
# perfoming the resampling() to see begining of the bussiness month 

bms= adaniports.resample(rule='BMS').sum()
bms
```

```python
# perfoming the resampling() to see weekly friday stock price 

weekly = adaniports.resample(rule='W-Fri').sum()
weekly
```

# Downsampling

Aggregating higher frequency data to lower frequency is called downsampling .The
data you’re aggregating doesn’t need to be fixed frequently; the desired frequency defines bin edges that are used to slice the time series into pieces to aggregate.

It is used to convert higher frequncy into the lower frequncy is known as downsampling 

for example :- higher frequence is like hours and when it could be converted into lower frequecny that is in minutes

basically its work on time series 

to perform downsampling and upsampling we can create a date_range lets see the below codes

```python
# creating a dates range 

date_range = pd.date_range('1/1/23',periods=12, freq='T')
date_range
```

```python
# making a series and giving a index as date_range
ser = pd.Series(np.arange(len(date_range)),index=date_range)
ser
```

```python
# checking the datatype
type(ser)
```

# closed{‘right’, ‘left’}, default None

Which side of bin interval is closed. The default is ‘left’ for all frequency offsets except for ‘M’, ‘A’, ‘Q’, ‘BM’, ‘BA’, ‘BQ’, and ‘W’ which all have a default of ‘right’.

Bydefault it is closed to left if we not mentined closed 



```python
# now performing a downsampling by default it closed on left bin time  interval

ser.resample('5min').sum()
```

```python
# now mentionong the closed ='right ' parameter

ser.resample('5min', closed = 'right',label='right').sum()
```

```python
The frequency you pass defines bin edges in five-minute increments. By default, the
left bin edge is inclusive, so the 00:00 value is included in the 00:00 to 00:15 interval.1 Passing closed='right'
```

```python
ser.resample('D').mean()
```

# closed ='left'  is a parameter  and label parameter

```python
# now performing resampling on the close = 'left' parameter

ser.resample('5min',closed='left').sum()
```

# offset parameter : timedelta, default None Adjust the resampled time labels.

By using loffset we can manipualte datetimeindex 

```python
# subtracting one second from the left edge

man = ser.resample('5min', offset='1s').sum()
man
```
